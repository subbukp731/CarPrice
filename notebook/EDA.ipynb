```python
df['CompanyName'] = df['CarName'].apply(lambda x : x.split(' ')[0])
```

```python
df['CompanyName']
```

```python
df = df.drop('CarName',axis = 1)
```

```python
def replace_name(a,b):
    df['CompanyName'].replace(a,b,inplace = True)
```

```python
replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')
```

```python
plt.subplot(1,2,1)
sns.distplot(df['price'])
plt.title('Car price distribution list')

plt.subplot(1,2,2)
sns.boxplot(df['price'])
plt.title('Car proice spread')
plt.show()
```

```python
plt.figure(figsize=(25,6))

plt.subplot(1,2,1)
sns.countplot(df['symboling'])

plt.subplot(1,2,2)
sns.boxplot(x = df['symboling'], y = df['price'])
```

```python
df1 = pd.DataFrame(df.groupby(['enginetype'])['price'].mean().sort_values())
df1.plot.bar()
```

```python
def scatter(x):
    plt.scatter(df[x],df['price'])
    plt.title(x+  ' vs price')
```
